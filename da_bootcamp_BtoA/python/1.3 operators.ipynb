### Operators

```python
## Arithametic operation
a = 10
b = 5
result = a + b
print(result)
```

```python
add_result = a + b  # addition
sub_result = a - b  # subtraction
mul_result = a * b  # multiplication
div_result = a / b  # division
flr_result = a // b # floor division (quotient)
mod_result = a % b  # modulus operation (remainder)
exp_result = a ** b # exponential operation

print(add_result)
print(sub_result)
print(mul_result)
print(div_result)
print(flr_result)
print(mod_result)
print(exp_result)
```

```python
## Comparison operators
## ==
a = 10
b = 10
a == b
```

```python
str1 = "Htet"
str2 = "Htet"

str1 == str2
```

```python
## Not equal !=
str1 != str1
```

```python
## Greater than
num1 = 45
num2 = 55
num1 > num2
```

```python
## Less than <
num1 < num2
```

```python
## Greater than or equal to
number1 = 45
number2 = 45
number1 >= number2
```

```python
## Less than or equal to
number1 <= number2
```

### Logical operators

```python
## AND operation
X = True
Y = True
result = X and Y
print(result)
```

```python
## AND operation
X = True
Y = False
result = X and Y
print(result)
```

```python
## AND operation
X = False
Y = False
result = X and Y
print(result)
```

```python
## OR operation
X = False
Y = False
result = X or Y
print(result)
```

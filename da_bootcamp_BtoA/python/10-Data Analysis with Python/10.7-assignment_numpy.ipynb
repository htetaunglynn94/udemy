2. Create a NumPy array of shape (5, 5) with random integers. Extract the elements on the border.

```python
import numpy as np

arr = np.random.randint(1, 100, size=(5,5))
print("Original array:")
print(arr)

print("Extract elements on border:")
top = arr[0, :]
bottom = arr[-1, :]
left = arr[1:-1, 0]
right = arr[1:-1, -1]
border = np.concatenate([top, left, bottom, right])
print(border)
```

### Assignment 3: Array Operations

1. Create two NumPy arrays of shape (3, 4) filled with random integers. Perform element-wise addition, subtraction, multiplication, and division.

```python
import numpy as np

arr1 = np.random.randint(1, 100, size=(3,4))
arr2 = np.random.randint(1, 100, size=(3,4))

print("Array 1:")
print(arr1)

print("Array 2:")
print(arr2)

# Addition
print("Array addition:")
print(arr1 + arr2)

# Subtraction
print("Array subtraction:")
print(arr1 - arr2)

# Multiplication
print("Array multiplication:")
print(arr1 * arr2)

# Division
print("Array division:")
print(np.round(arr1 / arr2, 1))
```

### Assignment 4: Statistical Operations

1. Create a NumPy array of shape (5, 5) filled with random integers. Compute the mean, median, standard deviation, and variance of the array.

```python
import numpy as np

arr = np.random.randint(1, 100, size=(5,5))
print("Original array:")
print(arr)


print(f"Mean value: {np.mean(arr)}")
print(f"Median value: {np.median(arr)}")
print(f"Standard deviation value: {np.std(arr)}")
print(f"Variance value: {np.var(arr)}")
```

2. Create a NumPy array of shape (3, 3) with values from 1 to 9. Normalize the array (i.e., scale the values to have a mean of 0 and a standard deviation of 1).

```python
import numpy as np

arr = np.random.randint(1, 10, size=(3,3))
print("Original array:")
print(arr)

mean = np.mean(arr)
stddev = np.std(arr)
normalized = (arr - mean) / stddev
print(f"After normalize: {normalized}")
```

### Assignment 6: Linear Algebra

1. Create a NumPy array of shape (3, 3) representing a matrix. Compute its determinant, inverse, and eigenvalues.

```python
import numpy as np

arr = np.random.randint(1, 10, size=(3,3))
print("Original array:")
print(arr)

# Calculation determinant
print("Determinant:")
print(np.linalg.det(arr))

# Calucation inverse
print("Inverse:")
print(np.linalg.inv(arr))

# Calculation eigenvalues
print("Eigenvalues:")
print(np.linalg.eigvals(arr))
```

### Assignment 9: Structured Arrays

1. Create a structured array with fields 'name' (string), 'age' (integer), and 'weight' (float). Add some data and sort the array by age.

```python
dtypes = [('name', 'U20'),    # ''U20' = unicode string (up to 20 characters)
          ('age', 'i4'),      # 'i4' = 32-bit integer
          ('weight', 'f4')]   # 'f4' = 32-bit float

arr = np.array([('Htet',24,61),
                 ('Aung',31,55),
                 ('Lynn',26,65)], dtype=dtypes)

print("Original array:")
print(arr)

print("Sort array:")
sorted = np.sort(arr, order='age')
print(sorted)

print("Check field names:")
print(arr.dtype.names)
```

2. Create a structured array with fields 'x' and 'y' (both integers). Add some data and compute the Euclidean distance between each pair of points.

```python
dtypes = np.dtype([('x', 'i4'),    # ''U20' = unicode string (up to 20 characters)
                   ('y', 'i4')])    # 'i4' = 32-bit integer

# 1D array
arr = np.array([(12,23),
                (22,71),
                (18,43),
                (30,78)], dtype=dtypes)

print("Check field names:")
print(arr.dtype.names)

print("Original array:")
print(arr)

# Euclidean distance calculation
# print(arr['x'][:, np.newaxis])
# print(arr['x'])
# print(arr['x'][:, np.newaxis] - arr['x'])
eu_dst = np.sqrt((arr['x'][:,np.newaxis] - arr['x'])**2 + (arr['y'][:, np.newaxis]- arr['y'])**2)
print("After distance calculation:")
print(eu_dst)
```

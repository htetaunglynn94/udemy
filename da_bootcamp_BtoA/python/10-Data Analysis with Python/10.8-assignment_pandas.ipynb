### Assignment 4: Data Aggregation

1. Create a Pandas DataFrame with 2 columns: 'Category' and 'Value'. Fill the 'Category' column with random categories ('A', 'B', 'C') and the 'Value' column with random integers. Group the DataFrame by 'Category' and compute the sum and mean of 'Value' for each category.

```python
import pandas as pd
import numpy as np

df = pd.DataFrame(columns=['Category','Value'])
df['Category'] = [np.random.choice(['A','B','C']) for _ in range(10)]
df['Value'] = [np.random.randint(1, 10) for _ in range(10)]
print("Original dataframe:")
display(df)

print("After groupby:")
df.groupby('Category')['Value'].agg(['sum','mean'])
```

### Assignment 6: Time Series Analysis

1. Create a Pandas DataFrame with a datetime index and one column filled with random integers. Resample the DataFrame to compute the monthly mean of the values.

```python
import pandas as pd
import numpy as np


df = pd.DataFrame(columns=['Date','Sales'])
df['Date'] = pd.date_range(start='2025-01-01', end='2025-12-31', freq='D')
df['Sales'] = np.random.randint(1, 100, size=len(df))
df.set_index('Date', inplace=True)
print("Original Dataframe:")
display(df)

monthly = df.resample('ME')['Sales'].mean() # groupby time
monthly.index = monthly.index.month_name()
monthly.index.name = 'Month'
print("Monthly mean DataFrame:")
display(monthly)
```

2. Create a Pandas DataFrame with a datetime index ranging from '2021-01-01' to '2021-12-31' and one column filled with random integers. Compute the rolling mean with a window of 7 days.

```python
import pandas as pd
import numpy as np


df = pd.DataFrame(columns=['Date','Sales'])
df['Date'] = pd.date_range(start='2021-01-01', end='2021-12-31', freq='D')
df['Sales'] = np.random.randint(1, 100, size=len(df))
df.set_index('Date', inplace=True)
print("Original Dataframe:")
display(df)

print("Rolling mean with windows of 7 days:")
display(df['Sales'].rolling(window=7).mean()) #     data
```

### Assignment 8: Pivot Tables

1. Create a Pandas DataFrame with columns 'Date', 'Category', and 'Value'. Create a pivot table to compute the sum of 'Value' for each 'Category' by 'Date'.

```python
import pandas as pd
import numpy as np

dates = pd.date_range(start='2025-01-01', end='2025-01-31')
df = pd.DataFrame({"Date": np.random.choice(dates, size=20),
                   'Category': np.random.choice(['A','B','C','D'], size=20),
                   'Value': np.random.randint(1, 100, size=20)})
print("Original dataframe:")
display(df)

print("Pivot-table:")
display(df.pivot_table(values='Value', columns='Category', index='Date', aggfunc='sum'))
```

```python
import pandas as pd
from random import randint, choice, uniform

dates = pd.date_range(start='2022-01-01', end='2025-12-31')
revenue = [round(uniform(1, 100),2) for _ in range(len(dates))]
df = pd.DataFrame({'Date': dates,
                   'Revenue': revenue})
df['Year'] = df['Date'].dt.year
df['Quarter'] = df['Date'].dt.quarter
df.replace({1:'Q1', 2:'Q2',3:'Q3',4:'Q4'}, inplace=True)
display(df)

df.pivot_table(index='Year', columns='Quarter', values='Revenue', aggfunc='mean')
```

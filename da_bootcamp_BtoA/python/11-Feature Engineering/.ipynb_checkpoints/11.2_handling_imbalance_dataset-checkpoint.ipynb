## Handling Imbalanced Dataset

1. Up Sampling
2. Down Sampling

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Set the random seed for reproducibility
np.random.seed(123)

# Create a dataframe with two classes
n_samples = 1000
class_0_ratio = 0.9
n_class_0 = int(n_samples * class_0_ratio)
n_class_1 = n_samples - n_class_0
```

```python
n_class_0,n_class_1
```

```python
## CREATE MY DATAFRAME WITH IMBALANCED DATASET
class_0 = pd.DataFrame({
    'feature_1': np.random.normal(loc=0, scale=1, size=n_class_0),
    'feature_2': np.random.normal(loc=0, scale=1, size=n_class_0),
    'target': [0] * n_class_0
})

class_1 = pd.DataFrame({
    'feature_1': np.random.normal(loc=2, scale=1, size=n_class_1),
    'feature_2': np.random.normal(loc=2, scale=1, size=n_class_1),
    'target': [1] * n_class_1
})
```

```python
df = pd.concat([class_0,class_1]).reset_index(drop=True)
```

```python
df.tail()
```

```python
df['target'].value_counts()
```

```python
## upsampling
df_minority = df[df['target'] == 1] # 100
df_majority = df[df['target'] == 0] # 900
```

```python
from sklearn.utils import resample

df_minority_upsampled = resample(df_minority,
                                 replace = True, # Sample With replacement
                                 n_samples = len(df_majority), # 900 samples
                                 random_state = 42)
```

```python
df_minority_upsampled.shape
```

```python
df_minority_upsampled.head()
```

```python
df_upsampled = pd.concat([df_majority, df_minority_upsampled])
```

```python
df_upsampled['target'].value_counts()
```

```python
fig, x = plt.subplots(1,2, figsize=(8,4))
x[0].scatter(df['feature_1'], df['feature_2'], c=df['target'])
x[1].scatter(df_upsampled['feature_1'], df_upsampled['feature_2'], 
             c=df_upsampled['target'])

x[0].set_title("Before upsampling")
x[1].set_title("After upsampling")
plt.suptitle("UPSAMPLING duplicates data points at same placees")
plt.tight_layout()
plt.show()
```

## Down Sampling

```python
import pandas as pd

# Set the random seed for reproducibility
np.random.seed(123)

# Create a dataframe with two classes
n_samples = 1000
class_0_ratio = 0.9
n_class_0 = int(n_samples * class_0_ratio)
n_class_1 = n_samples - n_class_0

class_0 = pd.DataFrame({
    'feature_1': np.random.normal(loc=0, scale=1, size=n_class_0),
    'feature_2': np.random.normal(loc=0, scale=1, size=n_class_0),
    'target': [0] * n_class_0
})

class_1 = pd.DataFrame({
    'feature_1': np.random.normal(loc=2, scale=1, size=n_class_1),
    'feature_2': np.random.normal(loc=2, scale=1, size=n_class_1),
    'target': [1] * n_class_1
})

df = pd.concat([class_0, class_1]).reset_index(drop=True)

# Check the class distribution
print(df['target'].value_counts())
```

```python
## downsampling
df_minority = df[df['target'] == 1] # 100
df_majority = df[df['target'] == 0] # 900
```

```python
from sklearn.utils import resample

df_majority_downsampled = resample(df_majority,
                                   replace = False, #Sample without replacement (reduce)
                                   n_samples = len(df_minority), # 900 samples
                                   random_state = 42)
```

```python
df_majority_downsampled.shape
```

```python
df_majority_downsampled.head()
```

```python
df_majority_downsampled.target.value_counts()
```

```python
df_downsampled = pd.concat([df_minority, df_majority_downsampled])
df_downsampled.head()
```

```python
df_downsampled['target'].value_counts()
```

NOTE: **Down sampling**
is not a good choice because we will loose lots of data.

### Variables

**Variables** are fundamental elements in programming used to store data that can be referenced manipulated in a program. In python, variables are created when you assign a value to them, and they do not need explicit declaration to reserve memory space.The declaration happens automatically when you assign a value to a variable.

```python
## Declaring and assigning variables
age = 32
height = 5.9
name = "Htet"
is_student = True

## Printing the variables
print('age:', age)
print('height:', height)
print('name:', name)
print(is_student)
```

```python
## Naming conventions
## Variable names should be descriptive
## They must start with a letter or an '_' and contain letter, numbers and underscores
## Variabels names are case sensitive

## Valid variable names
first_name = "Htet"
last_name = "Aung"
```

```python
## Invalid variable names
# 2age = 30
# first-name = "Htet"
# @name = "Aung"
```

```python
## Case sensitivity
name = "Htet"
Name = "Aung"
name == Name
```

```python
## Understanding variable types
## Python is dynamically typed; type of a variable is determined at runtime
age = 32 # int
height = 6.1 # float
name = "Htet" # string
is_student = True # bool

print(type(age))
print(type(height))
print(type(name))
print(type(is_student))
```

```python
## Type checking and conversion
print(type(age))
age_str = str(age)
print(age_str)
print(type(age_str))
```

```python
age = '22'
print(type(int(age)))
```

```python
## Cannot convert pure words to string
name = "Htet"
print(type(int(name)))
```

```python
height = 4.11
print(type(height))
print(type(int(height)))
print(int(height))
```

```python
## Dynamic typing
var = 10
print(var, type(var))

var = "Hello"
print(var, type(var))

var = 3.14
print(var, type(var))
```

```python
## Input
age = int(input("What is the age:")) # originally, it is string
print(age, type(age))
```

```python
## Simple calculator
num1 = float(input("Enter first number:"))
num2 = float(input("Enter second number:"))

sum = num1 + num2
dif = abs(num1 - num2)
pro = num1 * num2
quo = num1 / num2

print("Sum:", sum)
print("Difference:", dif)
print("Product:", pro)
print("Quotient:", quo)
```
